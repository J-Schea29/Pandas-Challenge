```python
#import pandas module
import pandas as pd
import os

#create path for files
csv1_path = os.path.join(".", "Resources", "schools_complete.csv")
csv2_path = os.path.join(".", "Resources", "students_complete.csv")
school_data = pd.read_csv(csv1_path)
student_data = pd.read_csv(csv2_path)
```

```python
#1. District Summary

#merge dataframes
merged_df = pd.merge(student_data, school_data, on= "school_name")
merged_df
```

```python
#count number of schools
school_count = school_data["school_name"].count()

#count number of students
student_count = merged_df["student_name"].count()

#add all budgets together
school_budget = school_data["budget"].sum()

#average math score
distr_average_math = round(merged_df["math_score"].mean(), 2)

#average reading score
distr_average_reading = round(merged_df["reading_score"].mean(), 2)

#print results(hopefully this is what project wants me to do)
print(f"The District has {school_count} Schools")
print(f"The District has {student_count} students")
print(f"District Budget: ${school_budget}")
print("--------------------------")
print(f"District Average Math Score: {distr_average_math}")
print(f"District Average Reading Score: {distr_average_reading}")

#count total passing and failing
reading_count = merged_df["reading_score"].count()
math_count = merged_df["math_score"].count()
passing_count = merged_df["Student ID"].count()

#locate tests passed in math, reading, and both
passing_reading = merged_df.loc[merged_df["reading_score"] >= 70, :]
passing_math = merged_df.loc[merged_df["math_score"] >= 70, :]
passing = merged_df.loc[(merged_df["math_score"] >= 70) & (merged_df["reading_score"] >= 70)]

#count just passing students
total_passing_read = passing_reading["reading_score"].count()
total_passing_math = passing_math["math_score"].count()
total_passing = passing["Student ID"].count()                         

#divide passing students by the original counts and convert to percents
dist_read_perc = round(int(total_passing_read)/int(reading_count) *int(100), 2)
dist_math_perc = round(int(total_passing_math)/int(math_count) *int(100), 2)
dist_pass_perc = round(int(total_passing)/int(passing_count) *int(100), 2)                         

#print result
print("-------------------------------------")
print(f"Percentage of Students passing reading: {dist_read_perc}%")
print(f"Percentage of Students passing math: {dist_math_perc}%")
print(f"Percentage of Students passing: {dist_pass_perc}%")
```

```python
#2. School Summary
student_data_df = student_data[["school_name", "reading_score", "math_score"]]
school_group = student_data_df.groupby(["school_name"])
aver_group = school_group.mean()
aver_group = aver_group.rename(columns ={"reading_score": "average_reading_score", "math_score": "average_math_score"})
```

```python
passing_reading2 = student_data_df.loc[merged_df["reading_score"] >= 70, :]
passing_math2 = student_data_df.loc[merged_df["math_score"] >= 70, :]
passing2 = student_data_df.loc[(merged_df["math_score"] >= 70) & (merged_df["reading_score"] >= 70)]

school_group2 = passing_reading2.groupby(["school_name"])
counted_group = school_group2.count()
counted_group

school_group3 = passing_math2.groupby(["school_name"])
counted_group2 = school_group3.count()
counted_group2

school_group4 = passing2.groupby(["school_name"])
counted_group3 = school_group4.count()
counted_group3

# percentages = counted_group[["school_name", "reading_score"]]
# percentages["passed_math"] = counted_group2["math_score"]
# percentages["passed"] = counted_group3["math_score"]
# percentages
merged1 = pd.merge(counted_group, counted_group2, on= "school_name")
merged2= pd.merge(merged1, counted_group3, on= "school_name")
del merged2["math_score_x"]
del merged2["reading_score_y"]
del merged2["math_score"]
merged2 = merged2.rename(columns ={"reading_score_x": "passed_reading", "math_score_y": "passed_math", "reading_score": "total"})
```

```python
import pandas as pd
import os

merge3 = pd.merge(school_data, aver_group, on= "school_name")
merged_data2 = pd.merge(merge3, merged2, on= "school_name")

merged_data2["money_per_student"] = merged_data2["budget"]/merged_data2["size"]
merged_data2["% Passed Reading"]= (merged_data2["passed_reading"]/merged_data2["size"]) *int(100)
merged_data2["% Passed Math"] = (merged_data2["passed_math"]/merged_data2["size"]) *int(100)
merged_data2["% Passed"] = (merged_data2["total"]/merged_data2["size"]) *int(100) 
del merged_data2["passed_reading"]
del merged_data2["passed_math"]
del merged_data2["total"]
merged_data2["budget"] = merged_data2["budget"].astype(float).map("${:,.2f}".format)
merged_data2["money_per_student"] = merged_data2["money_per_student"].astype(float).map("${:,.2f}".format)
merged_data2
```

```python
#3 Top 5 Performing Schools (By % Overall Passing)
top5 = merged_data2.sort_values("% Passed", ascending=False).head(5)
top5 
```

```python
#4 Bottom 5 Performing Schools (By % Overall Passing)
bottom5 = merged_data2.sort_values("% Passed", ascending=True).head(5)
bottom5
```

```python
#5 Math Scores by Grade*
grade_df= student_data[["grade", "school_name", "math_score"]]
grade_group = grade_df.groupby(["school_name", "grade"])
mean_grade = grade_group.mean()
pivot_grade = mean_grade.pivot_table("math_score", "school_name", "grade")
pivot_grade
```

```python
#6 Reading Scores by Grade
grade_df2= student_data[["grade", "school_name", "reading_score"]]
grade_group2 = grade_df2.groupby(["school_name", "grade"])
mean_grade2 = grade_group2.mean()
pivot_grade2 = mean_grade2.pivot_table("reading_score", "school_name", "grade")
pivot_grade2
```

```python
#7 Scores by School Spending 
spending = merged_data2[["money_per_student", "average_reading_score",
                      "average_math_score", "% Passed Reading", "% Passed Math", "% Passed"]]

spending['money_per_student'] = spending['money_per_student'].str.replace(',', '').str.replace('$', '').astype(float)

bins = [550, 600, 620, 640, 660]
spending_catergory = [4, 3, 2, 1]
spending["spending_per_student_catergory"] = pd.cut(spending["money_per_student"], bins, labels = spending_catergory)
spending_catergory = spending.groupby(["spending_per_student_catergory"])
mean_spending = spending_catergory.mean()
mean_spending["money_per_student"] = mean_spending["money_per_student"].astype(float).map("${:,.2f}".format)
mean_spending
```

```python
#8 Scores by School Size
sizes = merged_data2[["size", "average_reading_score",
                      "average_math_score", "% Passed Reading", "% Passed Math", "% Passed"]]
bins2 = [0, 1000, 3000, 10000]
types_size = ["small", "medium", "large"]
sizes["size_type"] = pd.cut(sizes["size"], bins2, labels = types_size)
size_type = sizes.groupby(["size_type"])
size_type.mean()
```

```python
#9 Scores by School Type
school_sorted = merged_data2[["type", "average_reading_score",
                            "average_math_score", "% Passed Reading", "% Passed Math", "% Passed"]]
school_sorted = school_sorted.groupby(["type"])   
school_type = school_sorted.mean()
school_type
```
